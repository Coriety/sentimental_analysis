**Loading a Pre-Trained Sentiment Analysis** **Model**

```python
from transformers import pipeline

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")
# Load the pre-trained DistilBERT model for sentiment analysis
sentiment_pipeline = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english",device=0 if device == "cuda" else -1)

# Load the corresponding tokenizer for the sentiment analysis model
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

**Performing Sentiment Analysis**

```python
# Function to classify sentiment in batches
def splitting(tokens, max_length=510):


  chunks=[tokens[i: i+max_length] for i in range(0, len(tokens), max_length)]
  return  chunks




def classify_sentiment(review):

  review = str(review)
  tokenized_output = tokenizer(review, truncation=False, padding=False, return_tensors='pt')
  tokens = tokenized_output["input_ids"][0]  # Get token IDs
  tokenized_length = len(tokens)

  if tokenized_length >510:
    token_chunks = splitting(tokens, max_length=510)
    review_chunks = [tokenizer.decode(chunk, skip_special_tokens=True) for chunk in token_chunks]
    chunk_sentiments = [sentiment_pipeline(chunk ,truncation=True)[0] for chunk in review_chunks]


    # Aggregate sentiment scores
    sentiment_scores = {"POSITIVE": 0, "NEGATIVE": 0, "NEUTRAL":0}
    for sentiment in chunk_sentiments:
        label = sentiment["label"]
        score = sentiment["score"]
        sentiment_scores[label] += score

    # Determine the final sentiment
    total_score = sum(sentiment_scores.values())

    # Normalize scores to compare percentages
    for key in sentiment_scores:
        sentiment_scores[key] /= total_score

    # **Classification Logic with Threshold**
    if sentiment_scores["POSITIVE"] > 0.5:
        final_sentiment = "POSITIVE"
    elif sentiment_scores["NEGATIVE"] > 0.5:
        final_sentiment = "NEGATIVE"
    else:
        final_sentiment = "NEUTRAL"

    return final_sentiment

  else:
    sentiment = sentiment_pipeline(review,truncation=True)[0]
    final_sentiment = sentiment["label"]

    return final_sentiment

# review="nice hotel expensive parking got good deal stay hotel anniversary, arrived late evening took advice previous reviews did valet parking, check quick easy, little disappointed non-existent view room room clean nice size, bed comfortable woke stiff neck high pillows, not soundproof like heard music room night morning loud bangs doors opening closing hear people talking hallway, maybe just noisy neighbors, aveda bath products nice, did not goldfish stay nice touch taken advantage staying longer, location great walking distance shopping, overall nice experience having pay 40 parking night,"
# print(classify_sentiment(review))

batch_size=50
df["adjusted_sentiment"] = pd.Series(dtype="object")
def process_reviews(batch):
    result= [classify_sentiment(review) for review in batch]
    torch.cuda.empty_cache()  # Free up GPU memory after each batch
    return result
batches=[df.iloc[i:i+batch_size]["Review"].tolist() for i in range(0, len(df), batch_size)]
results=Parallel(n_jobs=2, timeout=1000)(delayed(process_reviews)(batch) for batch in batches)
df["adjusted_sentiment"] = np.array([sentiment for batch in results for sentiment in batch], dtype=object)




# # Check sentiment distribution
print(df['adjusted_sentiment'].value_counts())



```

```python
# # Plot sentiment distribution
df['adjusted_sentiment'].value_counts().plot(kind='bar', color=['green', 'red'])
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```

```python
# Filter negative reviews
negative_reviews = df[df["adjusted_sentiment"] == "NEGATIVE"]["Review"]
print(f"Number of negative reviews: {len(negative_reviews)}")
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string

# Download NLTK resources
nltk.download("punkt")
nltk.download("stopwords")
nltk.download('punkt_tab')
```

```python
def preprocess_text(text):
  text = text.lower()
  text = text.translate(str.maketrans("", "", string.punctuation))
  tokens = word_tokenize(text)
  stop_words = set(stopwords.words("english"))
  filtered_tokens = [word for word in tokens if word not in stop_words]
  return filtered_tokens
negative_tokens = negative_reviews.apply(preprocess_text)
```

**TOPIC MODELLING-Latent Dirichlet Allocation (LDA).**

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
negative_texts = [" ".join(tokens) for tokens in negative_tokens]
vectorizer = CountVectorizer(stop_words="english", max_features=1000)
doc_term_matrix = vectorizer.fit_transform(negative_texts)
lda = LatentDirichletAllocation(n_components=100, random_state=42)
lda.fit(doc_term_matrix)
# Display topics
negative_keywords = {"bad", "terrible", "awful", "worst", "disgusting", "horrible", "poor", "dirty", "noisy", "rude"}
words = vectorizer.get_feature_names_out()
negative_topics = []
for topic_idx, topic in enumerate(lda.components_):
    top_words = [words[i] for i in topic.argsort()[-10:]]  # Top 5 words
    if any(word in negative_keywords for word in top_words):
        negative_topics.extend(top_words)


word_freq = {word: negative_topics.count(word) for word in set(negative_topics)}
wordcloud = WordCloud(width=800, height=400, background_color="black", colormap="Reds").generate_from_frequencies(word_freq)

# Step 7: Plot the Word Cloud
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("🔹 Negative Topics from Reviews (LDA)")
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Use CountVectorizer to find bigrams
bigram_vectorizer = CountVectorizer(ngram_range=(2, 2), stop_words="english")
bigram_counts = bigram_vectorizer.fit_transform(negative_reviews)
bigram_freq = dict(zip(bigram_vectorizer.get_feature_names_out(), bigram_counts.toarray().sum(axis=0)))

# Sort by frequency
sorted_bigram_freq = sorted(bigram_freq.items(), key=lambda x: x[1], reverse=True)

# Display top 10 most frequent bigrams
print("\nTop 10 most frequent bigrams:")
for bigram, freq in sorted_bigram_freq[:1000]:
    print(f"{bigram}: {freq}")
```
